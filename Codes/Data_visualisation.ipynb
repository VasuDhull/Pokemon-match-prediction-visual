```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
path2  = '/CSV Files/'
path = os.path.dirname(os.getcwd())
```

```python
pokemon = pd.read_csv(path + path2 +'pokemon_data.csv')
combats = pd.read_csv(path + path2 +'combats.csv')
```

```python
types = pokemon['Type 1'].unique()
```

```python
print("The diffrent types of pokemon are :")
types
```

```python
type_sum = pokemon['Type 1'].value_counts() + pokemon['Type 2'].dropna().value_counts()
type_sum
```

```python
bright_colors = [
    '#FF0000',  # Red
    '#FFA500',  # Orange
    '#FFFF00',  # Yellow
    '#00FF00',  # Lime
    '#00FFFF',  # Cyan
    '#0000FF',  # Blue
    '#800080',  # Purple
    '#FF1493',  # Deep Pink
    '#FFD700',  # Gold
    '#32CD32',  # Lime Green
    '#00CED1',  # Dark Turquoise
    '#9370DB',  # Medium Purple
    '#DC143C',  # Crimson
    '#48D1CC',  # Medium Turquoise
    '#FF6347',  # Tomato
    '#FF4500',  # Orange Red
    '#8A2BE2',  # Blue Violet
    '#00FF7F'   # Spring Green
]
```

```python
fig, ax = plt.subplots(figsize=(16, 6))
plt.bar(types, type_sum, color=bright_colors)

# Add labels and title
plt.xlabel('Types')
plt.ylabel('Type Count')
plt.title('Pokemon Types with their values')

# Display the bar graph
plt.show()
```
