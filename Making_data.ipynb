```python
import pandas as pd
import numpy as np
```

```python
pokemon = pd.read_csv('pokemon_data.csv')
combats = pd.read_csv('combats.csv')
```

```python
pokemon
```

```python
type_1_list=pokemon['Type 1']
type_2_list=pokemon['Type 2']

type_1_list=pd.get_dummies(
    type_1_list,
    prefix=None,
    prefix_sep='_', 
    dummy_na=True, 
    columns=['Type 1'], 
    sparse=False, 
    drop_first=False, 
    dtype=int
)

type_2_list=pd.get_dummies(
    type_2_list,
    prefix=None,
    prefix_sep='_', 
    dummy_na=True, 
    columns=['Type 2'], 
    sparse=False, 
    drop_first=False, 
    dtype=int
)
all_type_list = type_1_list + type_2_list
all_type_list
```

```python
combined_dataset = pokemon.join(all_type_list)
```

```python
combined_dataset
```

```python
combined_dataset.drop(np.NaN, axis = 1, inplace= True)
```

```python
combined_dataset
```

```python
combined_dataset.pop('Type 1')
combined_dataset.pop('Type 2')
combined_dataset.pop('Name')
combined_dataset.pop('Generation')
combined_dataset.pop('Legendary')
```

```python
combined_dataset
```

```python
def get_features(pokemon_id,pokemon):
    return (pokemon.iloc[pokemon_id - 1, 0::1])
```

```python
get_features(23, pokemon)
```

```python
def rename_column(first_or_second, features):
     list_ = features.columns.to_list()
     if first_or_second:
          for i in range(len(list_)):
               list_[i] = 'First_' + list_[i] 

     else:
          for i in range(len(list_)):
               list_[i] = 'Second_' + list_[i] 

     features.columns = list_
```

```python
first_pokemon = combined_dataset.copy()
rename_column(True, first_pokemon)

list_first = combats.iloc[:, 0].to_list()
temp_ = list()

for i in range(len(list_first)):
    temp_.append(get_features(list_first[i], first_pokemon).to_frame().T)

df_first_pokemon = pd.concat(temp_, ignore_index=True)
df_first_pokemon
```

```python
temp_.clear()
second_pokemon = combined_dataset.copy()
rename_column(False, second_pokemon)

list_second = combats.iloc[:, 1].to_list()


for i in range(len(list_second)):
    temp_.append(get_features(list_second[i], second_pokemon).to_frame().T)

df_second_pokemon = pd.concat(temp_, ignore_index=True)
df_second_pokemon
```

```python
df_first_pokemon.drop('First_#', axis = 1, inplace=True)
df_second_pokemon.drop('Second_#', axis = 1, inplace=True)
```
